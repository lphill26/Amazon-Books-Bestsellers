```python
# https://www.kaggle.com/datasets/obaidhere/amazon-bestsellers-price-vs-user-rating-analysis

#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data
df = pd.read_csv('bestsellers.csv')
df.head()
```

```python
#data information
df.info()
```

```python
#data statistics
df.describe()
```

```python
#missing values
df.isna().sum()
```

```python
#duplicates
df.duplicated().sum()
```

```python
#percentile distribution of the user rating numerical column
df['User Rating'].quantile([0.25, 0.5, 0.75,])
```

```python
#percentile distribution of the reviews numerical column
df['Reviews'].quantile([0.25, 0.5, 0.75,])
```

```python
#percentile distribution of the price numerical column
df['Price'].quantile([0.25, 0.5, 0.75,])
```

```python
#percentile distribution of the year numerical column
df['Year'].quantile([0.25, 0.5, 0.75,])
```

```python
#name categorial column analysis
df['Name'].value_counts()
df['Name'].nunique()
```

```python
#author categorial column analysis
df['Author'].value_counts()
df['Author'].nunique()
```

```python
#genre categorial column analysis
df['Genre'].value_counts()
df['Genre'].nunique()
```

```python
#grouping the best selling book by the years
df.groupby('Name')['Year'].agg(['mean', 'median', 'count'])
```

```python
#grouping the best selling book names by the price
df.groupby('Name')['Price'].agg(['mean', 'median', 'count'])
```

```python
#grouping the best selling book names by the reviews
df.groupby('Name')['Reviews'].agg(['mean', 'median', 'count'])
```

```python
#grouping the best selling book names by the user ratings
df.groupby('Name')['User Rating'].agg(['mean', 'median', 'count'])
```

```python
#grouping the genre types by user ratings
df.groupby('Genre')['User Rating'].agg(['mean', 'median', 'count'])
```

```python
#grouping the genre types by reviews
df.groupby('Genre')['Reviews'].agg(['mean', 'median', 'count'])
```

```python
#grouping the genre types by price
df.groupby('Genre')['Price'].agg(['mean', 'median', 'count'])
```

```python
#grouping the genre types by year
df.groupby('Genre')['Year'].agg(['mean', 'median', 'count'])
```

```python
#asociation between best selling books and their genres
pd.crosstab(df['Name'], df['Genre'])
```

```python
#histogram of the number of reviews per book
df['Reviews'].hist()
```

```python
#bar chart of genre type of book
df['Genre'].value_counts().plot(kind='bar')
```

```python
#boxplot grouped by the genre and price of a the best selling books
df.boxplot(column='Price', by='Genre')
```

```python
#scatter plot of the user ratings and price
df.plot.scatter(x='User Rating', y='Price')
#the lower the price of book the high user rating
```

```python
#Correlation Heat Map
df.corr(numeric_only=True).style.background_gradient(cmap='coolwarm')
```
